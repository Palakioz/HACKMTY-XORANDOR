```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data2/credit_card_transactions.csv')
df.drop(['Unnamed: 0'], axis=1, inplace=True)
df.head()
```

```python
df.info()
```

# Data Cleanup

```python
# Convert 'trans_date_trans_time' and 'dob' columns to datetime format
df['trans_date_trans_time'] = pd.to_datetime(df['trans_date_trans_time'])
df['dob'] = pd.to_datetime(df['dob'])

# Calculate age
current_date = df['trans_date_trans_time'].max()
df['age'] = (current_date - df['dob']).dt.days // 365

# Filter transactions for individuals between 18-25 years old
filtered_df = df[(df['age'] >= 18) & (df['age'] <= 25)]
filtered_df.head()
```

```python
# We want non-fraudulent transactions

filtered_df = filtered_df[filtered_df['is_fraud'] == 0]
filtered_df.head()
```

```python
# Drop irrelevant columns

filtered_df = filtered_df.drop(columns=['first', 'last', 'cc_num', 'merchant', 'street', 'city', 'state', 'zip', 'lat', 'long', 'trans_num', 'unix_time', 'merch_lat', 'merch_long', 'is_fraud', 'merch_zipcode', 'dob', 'trans_date_trans_time'])
```

```python
filtered_df.info()
```

# Feature Engineering

```python
object_columns = filtered_df.select_dtypes(include='object').columns
for column in object_columns:
    unique_values = filtered_df[column].nunique()
    print("------------------")
    print(f"Unique values for {column}: {unique_values}")
```

```python
filtered_df.head()
```

```python
resumen = filtered_df.groupby(['gender', 'age', 'category'])['amt'].agg(['sum', 'count']).reset_index()
resumen.rename(columns={'sum': 'total_amount', 'count': 'total_transactions'}, inplace=True)
resumen.to_csv('data2/credit_card_transactions_summary.csv', index=False)
```
